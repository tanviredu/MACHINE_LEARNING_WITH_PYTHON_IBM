```python
#### load partial images
import cv2
import os
#### this function will import the data from the folder
def load__partial_images_from_folder(folder,target):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append([img,target])
    return images[:10]
```

```python
blb = load__partial_images_from_folder('../input/Bacterial leaf blight/',0)
```

```python
bs = load__partial_images_from_folder('../input/Brown spot/',1)
```

```python
ls = load__partial_images_from_folder('../input/Leaf smut/',2)
```

```python
blb.extend(bs)
```

```python
blb.extend(ls)
```

```python
training_data = blb
```

```python
feature_matrix = []
target = []
for x,y in training_data:
    feature_matrix.append(x)
    target.append(y)
    
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=[]
IMG_SIZE= 32
for x in feature_matrix:
    new_array = cv2.resize(x,(IMG_SIZE,IMG_SIZE))
    X.append(new_array)
```

```python
Xx = []
for x in X:
    tmp = x/255
    Xx.append(tmp)
```

```python
x_train,x_test,y_train,y_test = train_test_split(Xx,target)
```

```python
import sys
from matplotlib import pyplot
from keras.datasets import cifar10
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Dense
from keras.layers import Flatten
from keras.optimizers import SGD


def define_model():
	model = Sequential()
	model.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same', input_shape=(32, 32, 3)))
	model.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'))
	model.add(MaxPooling2D((2, 2)))
	model.add(Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'))
	model.add(Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'))
	model.add(MaxPooling2D((2, 2)))
	model.add(Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'))
	model.add(Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'))
	model.add(MaxPooling2D((2, 2)))
	model.add(Flatten())
	model.add(Dense(128, activation='relu', kernel_initializer='he_uniform'))
	model.add(Dense(3, activation='softmax'))
	# compile model
	opt = SGD(lr=0.001, momentum=0.9)
	model.compile(optimizer=opt, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
	return model
```

```python
model =  define_model()
history = model.fit(np.array(x_train),np.array(y_train),epochs=50)
```

```python
from keras.utils.vis_utils import plot_model
```

```python
plot_model(model)
```

```python
print(history.history.keys())
# summarize history for accuracy
plt.plot(history.history['accuracy'])
#plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
#plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
predicted = model.predict(np.array(x_test))
```

```python
result = []
for item in predicted:
    result.append(np.argmax(item))
```
