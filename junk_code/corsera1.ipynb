```
### binary classification

dp1 = {'partno':100,'maxtemp':35,'mintemp':12,'maxvibration':12,'broken':0}
dp2 = {'partno':101,'maxtemp':46,'mintemp':35,'maxvibration':21,'broken':0}
dp3 = {'partno':130,'maxtemp':56,'mintemp':46,'maxvibration':3412,'broken':1}
dp4 = {'partno':131,'maxtemp':58,'mintemp':48,'maxvibration':3542,'broken':1}
```

```
#### using in the logistic regression
## fist make a sigmoid function
import math

def sigmoid(x):
  return 1/(1+math.exp(-x))
```

```
## with sigmoid function

def mlpredict(dp):
  return 1 if sigmoid(w1+w2*dp['maxtemp']+w3*dp['mintemp']+w4*dp['maxvibration']) > .7 else 0
```

```
print (mlpredict(dp1))
print (mlpredict(dp2))
print (mlpredict(dp3))
print (mlpredict(dp4))
```
