```
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
```

```
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```
## setting parameter for random forest

## first number of feature is 784. how is that?
## its a 2D array and every element of the array is a 
## feature so it has 28*28 = 784
## and the target is 10 0-9



## tensorflow random forest take 6 parameter

epoch_size = 100  ## how many times trained
batch_size = 1024 ## number of samples per batch
num_class = 10   ## target class
num_feature = 784 ## number of feature
num_trees = 500; ## how many random trees 
max_nodes = 10000;  ## how many nodes every tree have



```

```
## like the tensorflow we need two placeholder
X = tf.placeholder(tf.float32,[None,28,28])
Y = tf.placeholder(tf.float32,[None])
```

```
from tensorflow.contrib.tensor_forest.python import tensor_forest
```

```
hparams = tensor_forest.ForestHParams(num_classes=num_class, num_features=num_feature, num_trees=num_trees, max_nodes=max_nodes).fill()
```

```
## creating the graph
forest_graph = tensor_forest.RandomForestGraphs(hparams)
```

```
## train with labels and find the loss like the model.score()
train_op = forest_graph.training_graph(X,Y)
```

```
## find the loss
loss_op = forest_graph.training_loss(X, Y)
```

```
infer_op, _, _ = forest_graph.inference_graph(X) ## its actually accuracy
## but we dont feed the data yest thats why called it inference
```

```
correct_prediction = tf.equal(tf.argmax(infer_op, 1), tf.cast(Y, tf.int64))

# second parameter in the argmax is the axis 
# this cell not only find one prediction acuracy it find all the accuracy
## tf equal is giving the the 0 and 1 is the prediction is false and correct
```

```

accuracy_op = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

## this is optimization
```

```
## initialize the variable
init  = tf.global_variables_initializer()
```

```
with tf.Session() as sess:
  sess.run(init)
  for item in range(epoch_size):
    acc = sess.run(accuracy_op, feed_dict={X: x_train, Y: y_train})
  
```
